```python
from kornia.color import hsv_to_rgb,rgb_to_hsv
from PIL import Image
from torchvision import transforms
import torch
img = Image.open(r'/root/autodl-tmp/RandStainNA-master/segmentation/choose25/TCGA-2Z-A9J9-01A-01-TS1_0_512.png').convert('RGB')
transform1 = transforms.Compose([transforms.ToTensor()])
img0 = transform1(img)

img1 = img0.unsqueeze(0)
print(img1)
xx = rgb_to_hsv(img1)
print(xx)
img_h = xx[0,0]
img_s = xx[0,1]
img_v = xx[0,2]
xxx = hsv_to_rgb(xx)
print(xxx)
h_mean = torch.mean(img_h)
s_mean = torch.mean(img_s)
v_mean = torch.mean(img_v)

h_var = torch.var(img_h, unbiased=False) 
s_var = torch.var(img_s, unbiased=False) 
v_var = torch.var(img_v, unbiased=False) 
print(h_mean,s_mean,v_mean,h_var,s_var,v_var)
```

```python
from hed import hed_to_rgb,rgb_to_hed#9.11
from PIL import Image
from torchvision import transforms
import torch
img = Image.open(r'/root/autodl-tmp/RandStainNA-master/segmentation/choose25/TCGA-2Z-A9J9-01A-01-TS1_0_512.png').convert('RGB')
transform1 = transforms.Compose([transforms.ToTensor()])
img0 = transform1(img)

img1 = img0.unsqueeze(0)
print(img1)
xx = rgb_to_hed(img1)
print(xx)
img_h = xx[0,0]
img_e = xx[0,1]
img_d = xx[0,2]
xxx = hed_to_rgb(xx)
print(xxx)
h_mean = torch.mean(img_h)
e_mean = torch.mean(img_e)
d_mean = torch.mean(img_d)

h_var = torch.var(img_h, unbiased=False) 
e_var = torch.var(img_e, unbiased=False) 
d_var = torch.var(img_d, unbiased=False) 
print(h_mean,e_mean,d_mean,h_var,e_var,d_var)
```

```python
##深度学习过程中，需要制作训练集和验证集、测试集。

import os, random, shutil


def moveFile(fileDir):
    pathDir = os.listdir(fileDir)  # 取图片的原始路径
    filenumber = len(pathDir)
    rate = 0.2  # 自定义抽取图片的比例，比方说100张抽10张，那就是0.1
    picknumber = int(filenumber * rate)  # 按照rate比例从文件夹中取一定数量图片
    sample = random.sample(pathDir, picknumber)  # 随机选取picknumber数量的样本图片
    print(sample)
    for name in sample:
        shutil.move(fileDir + name, tarDir + name)
    return


if __name__ == '__main__':
    fileDir = "/root/autodl-tmp/RandStainNA-master/segmentation/train_split/train/bound/"  # 源图片文件夹路径 
    tarDir = '/root/autodl-tmp/RandStainNA-master/segmentation/train_split/test/bound/'  # 移动到新的文件夹路径
    moveFile(fileDir)
    
```
